```python
import matplotlib.pyplot as plt
import numpy as np
#Creating and plotting a Linear Equation and Quadratic Equation with Numpy and Matplotlib
# Linear Equation --> Y=MX+C where M=Slope C is the Y-Intercept
# Quadratic Equation --> Y=AX^2+BX+C where A,B,C are Constants 
```

```python
#Creating an Array using x values
x=np.linspace(start=-10, stop=10, num=100)
print(x)
```

```python
#Plotting a Linear Equation
m=4
c=3
y=m*x+c
#plotting the Equation
plt.plot(x,y,label='y=4x+3')
plt.title("Plot of the Linear Equation") #Giving Title to our Plot
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,10) #lower limit, upper limit
plt.ylim(0,20) #lower limit, upper limit
plt.legend() #It shows the Label at the top of Graph
plt.grid(True)
plt.show()
```

```python
#plotting a Quadratic Equation
#Definig Coefficients
a = 1
b = -5
c = 6
# Generating X-values
x = np.linspace(-1,7,200)
#Calculating Y values using the Quadratic Equation
y = a*x**2+b*x+c
plt.plot(x,y, label='y = x**2+(-5)*x+6') #Creating plot
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10,15) #lower limit, upper limit
plt.ylim(0,30) #lower limit, upper limit
plt.legend() #It shows the Label at the top of Graph
plt.grid(True)#Adding the Grid
plt.show()
```

```python
#Matplotlib Figure Object
fig = plt.figure() #It will create a figure object
#Adding the axes to the Figure(Left,Bottom,Width, Height)
ax = fig.add_axes([0.1,0.1,0.5,0.5])
plt.show()

          
```

```python
#Creating an array of X values
x =np.linspace(0,10,100)
#Calculating Y values based on X
y= np.sin(x)
#Creating a figure with specific size
fig = plt.figure(figsize=(4,4))
#Add an axes to the figure
ax= fig.add_axes([0.1,0.1,0.8,0.8])
#Plotiing the Figure
plt.plot(x,y, label='sin(x)',color='Red')
#Giving the title
plt.title("Simple Plot of Sinx")
plt.xlabel("X") #Defing X-Label
plt.ylabel("Y") #Defing Y-Label
plt.legend()
plt.show()
```

```python
#Example Code for  Linear Functions in Different Axes Positions

#Generating X values
x = np.linspace(0,10,100)

#Create a Blank canvas
fig = plt.figure(figsize=(10,6))

#Center Plot-Main Plot
axes_center= fig.add_axes([0.3,0.3,0.4,0.4]) #Left,Bottom, Height, Width
plt.plot(x,x+1)
plt.title('Center Plot:y=x+1')

#Top Plot - Summary
axes_top= fig.add_axes([0.3,0.75,0.4,0.2])
plt.plot(x,2*x+1)
plt.title('Top Plot:y=2*x+1')

#Right Plot - Auxiliary Data
axes_right= fig.add_axes([0.75,0.3,0.2,0.4])
plt.plot(x,3*x+1)
plt.title('Right Ploty=3*x+1')

#Left Plot - Contextual Information
axes_left= fig.add_axes([0.05,0.3,0.2,0.4])
plt.plot(x,3*x+1)
plt.title('Left Plot:y=3*x+1')

#Bottom Plot - Comparitive Data
axes_bottom= fig.add_axes([0.3,0.05,0.4,0.2])
plt.plot(x,-x+1)
plt.title('Bottom Plot:y=-x+1')

#Top Right - Highlight
axes_top_right= fig.add_axes([0.75,0.75,0.2,0.2])
plt.plot(x,4*x+1)
plt.title('Top_Right Plot:y=4*x+1')

plt.show()






```

```python
#Matplotlib Subplots:
#Creating a Figure and a single Subplot
fig ,ax =plt.subplots()
#plotting the Data
plt.plot([1,2,3,4,5],[1,4,9,16,25],label="y=x**2")
plt.legend()
plt.show()
```

```python
#Plotting the subplots
fig, axs= plt.subplots(2,2)    #axs=Axes
axs[0,0].plot([1,2,4,5,3],[1,4,16,5,9],label="y=x**2")#Top Left
axs[0,1].plot([1,4,8,6,5],[2,8,16,12,10],label="y=2*x") #Top Right
axs[1,0].plot([1,6,4,2,10],[0,3,2,1,5],label="y=x/2") #Bottom Left
axs[1,1].plot([1,2,4,5,6],[1,8,64,125,216],label='y=x**3') #Bottom Right
for ax in axs.flat:  #This expression returns a flattened iterator over all Axes objects in the axs array.
    ax.legend()
#Showing the plot
plt.show()
```

```python
#Plt.tight_layout()-It is used to adjust subplots parameters to give specified padding and avoid overlap between subplots.
fig, axs= plt.subplots(2,2, figsize=(8,8), dpi=200, sharex=True, subplot_kw={'facecolor':'lightblue'})    
axs[0,0].plot([1,2,4,5,3],[1,4,16,5,9],label="y=x**2")#Top Left
axs[0,1].plot([1,4,8,6,5],[2,8,16,12,10],label="y=2*x") #Top Right
axs[1,0].plot([1,6,4,2,10],[0,3,2,1,5],label="y=x/2") #Bottom Left
axs[1,1].plot([1,2,4,5,6],[1,8,64,125,216],label='y=x**3') #Bottom Right
plt.tight_layout() #Auto Adjustment of a image
plt.show()
```

```python
#Matplotlib Bar Graphs

categories = ["Apple","Mango","Grapes"]
values= [10,20,30]
#Creating Bar Graph
plt.bar(categories,values)
#Adding Titles
plt.title("Basic Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Horizontal Bar Graph

categories = ["Apple","Mango","Grapes"]
values= [10,20,30]
#Creating Horizontal Bar Graph
plt.barh(categories,values,color='r')  #barh - For creating Horizontal Graph
#Adding Titles
plt.title("Horizontal Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Bars with Different colors and Borders


categories = ["Apple","Mango","Grapes"]
values= [10,20,30]
#Creating Bar Graph
plt.bar(categories,values,color=['r','b','g'],edgecolor='y')  

plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Creation of stacked Bars

categories = ["Apple","Mango","Grapes"]
value1= [10,20,30]
value2= [2,3,5]
#Creating Bar Graph
plt.bar(categories,value1,label='Part-1')  
plt.bar(categories,value2,bottom=value1,label='Part-2')
plt.legend()

plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Creating Grouped Bar Graphs

#Creation of stacked Bars

categories = ["Apple","Mango","Grapes"]
value1= [10,20,30]
value2= [2,3,5]
x=np.arange(len(categories)) #Label Locations
width=0.35
#Creating Bar Graph
plt.bar(x-width/2,value1,width,label='Part-1')  
plt.bar(x+width/2,value2,width,label='Part-2')
plt.legend()
plt.title("Grouped Bar Graphs")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(x,categories)
plt.show()
```

```python
#Creating Grouped Bar Graphs

#Creation of stacked Bars

categories = ["Apple","Mango","Grapes"]
value1= [10,20,30]
value2= [2,3,5]
x=np.arange(len(categories)) #Label Locations
width=0.35
#Creating Bar Graph
plt.bar(x-width/2,value1,width,label='Part-1')  
plt.bar(x+width,value2,width,label='Part-2')   #This Line makes difference from thr above graph
plt.legend()
plt.title("Grouped Bar Graphs")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(x,categories)
plt.show()
```

```python
#Adding Error Bars

#Creation of stacked Bars

categories = ["Apple","Mango","Grapes"]
values= [10,20,30]
errors= [2,3,5]
#Creating Bar Graph
#xerr= It shows error bars in horizontally
plt.bar(categories,values,xerr=errors,capsize=6)  #Capsize - It specifies the length of the error bar caps.
plt.title("Bar Graph with Error bars")            


plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Creation of stacked Bars

categories = ["Apple","Mango","Grapes"]
values= [10,20,30]
errors= [2,3,5]
#Creating Bar Graph
#yerr= It shows error bars in vertically
plt.bar(categories,values,yerr=errors,capsize=10) #Capsize- It specifies the length of the error bar caps.
plt.title("Bar Graph with Error bars")            


plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
